```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Reading the Car Prices data

cars_info = pd.read_csv("CarPrices.csv")
cars_info
```

```python
cars_info.describe()
```

```python
# Checking the dataset for any null values

cars_info.isnull().sum()
```

```python
# Checking for any correlation between the columns

cars_info.corr()
```

```python
# The higest correaltion that we can see is between "enginesize" and "price"
# Let's plot it to see a better visualization

plt.figure(figsize = (14, 7))
sns.scatterplot(x = 'price', y = 'enginesize', data = cars_info)
```

```python
# Finding how many unique elements are in each column

for car in cars_info:
    print(f"{car} - {cars_info[car].nunique()}")
```

```python
# Before we can model our data, we have to convert the strings to numbers.

# Looking at the unique values above, we can drop the following columns
cars_info.drop(['car_ID','CarName','fuelsystem','enginetype'],axis=1,inplace=True)

# And now we can now convert the strings to numbers as shown below
cars_info['fueltype'] = cars_info['fueltype'].map({'gas':'0','diesel':'1'})
cars_info['aspiration'] = cars_info['aspiration'].map({'std':'0','turbo':'1'})
cars_info['doornumber'] = cars_info['doornumber'].map({'two':'2','four':'4'})
cars_info['carbody'] = cars_info['carbody'].map({'convertible':'0','hatchback':'1','sedan':'2','wagon':'3','hardtop':'4'})
cars_info['drivewheel'] = cars_info['drivewheel'].map({'rwd':'0','fwd':'1','4wd':'2'})
cars_info['enginelocation'] = cars_info['enginelocation'].map({'front':'0','rear':'1'})
cars_info['cylindernumber'] = cars_info['cylindernumber'].map({'four':'4','six':'6','five':'5','three':'3','twelve':'12','two':'2','eight':'8'})

cars_info.head()
```

```python
# From the scatter plot we saw earlier, we can say that "price" increases as "enginesize" increases


# train test split
from sklearn.model_selection import train_test_split

X = cars_info.drop('price',axis=1)
y = cars_info['price']

X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=55)

# fit 
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)

# predict
prediction = lr.predict(X_test)
```

```python
# Now let's test the accuracy of our prediction that "price" increases as "enginesize" increases

from sklearn.metrics import r2_score
print(r2_score(y_test,prediction))
```

Our prediction that "price" increases as "enginesize" increases has an accuracy of about 78%
